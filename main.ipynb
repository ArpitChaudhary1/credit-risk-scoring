# Loan Approval and Rejection 

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(r"E:\codes\AI&ML\project\credit risk scoring system\train_lending_club.csv")
data.head()
```

```python
cors = data.corr(numeric_only=True)
sns.heatmap(cors, annot=True)
```

```python
le = LabelEncoder()
data['sub_grade'] = le.fit_transform(data['sub_grade'])
```

```python
data.drop(columns=['fico_range_high'], inplace=True)
data.head()
```

```python
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
```

```python
data.drop(columns=['sub_grade' ,'installment'], inplace=True)
```

```python
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
```

```python
data.head()
```

```python
data['term'] = data['term'].apply(lambda x: int(x.split()[0]))
data.head()
```

```python
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
```

```python
data.isnull().sum()
```

```python
data.select_dtypes(include=['object']).nunique()
```

```python
data.drop(columns=['issue_d','addr_state'], inplace=True)
data.head()
```

encoded the rest useful data to train the model

```python
dummies = pd.get_dummies(data[['purpose','home_ownership','verification_status','application_type','initial_list_status']] , drop_first=True , dtype=int)
data = pd.concat([data,dummies], axis=1)
data.drop(columns=['purpose','home_ownership','verification_status','application_type','initial_list_status'], inplace=True)
```
